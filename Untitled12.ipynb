```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# 1. ë°ì´í„°ë² ì´ìŠ¤ ì„¤ì • (ì¡°ì‚¬í•˜ì‹  ìë£Œ ê¸°ë°˜)
drug_db = {
    "ì•„ì„¸íŠ¸ì•„ë¯¸ë…¸íœ": {
        "ìƒí’ˆëª…": ["íƒ€ì´ë ˆë†€", "íƒ€ì„¸ë†€", "ê²Œë³´ë¦°", "íœì˜"],
        "ì£¼ì˜ì‹ì¬ë£Œ": ["ë¹„íƒ€ë¯¼C", "ì‹ì´ì„¬ìœ ", "ì˜¤íŠ¸ë°€", "ì‹œë¦¬ì–¼", "ì—í‚¤ë‚˜ì‹œì•„"]
    },
    "ì´ë¶€í”„ë¡œíœ": {
        "ìƒí’ˆëª…": ["ì• ë“œë¹Œ", "ì´ì§€ì—”6 ì• ë‹ˆ", "ë¶€ë£¨íœ", "ê·¸ë‚ ì—”"],
        "ì£¼ì˜ì‹ì¬ë£Œ": ["ì¹´í˜ì¸", "ë§ˆëŠ˜", "ìƒê°•"]
    },
    "ë‚˜í”„ë¡ì„¼": {
        "ìƒí’ˆëª…": ["íƒì„¼", "ì´ì§€ì—”6 ìŠ¤íŠ¸ë¡±", "ë‚™ì„¼", "í˜ì¸ì—”ì ¤ ì„¼"],
        "ì£¼ì˜ì‹ì¬ë£Œ": ["ì¹´í˜ì¸", "ë§ˆëŠ˜", "ìƒê°•", "ì¸ì‚¼", "ì€í–‰ì—½"]
    }
}

reasons = {
    "ë¹„íƒ€ë¯¼C": "ë…ì„± ì¦ê°€ ìœ„í—˜(3g ì´ìƒ ê¸ˆì§€)",
    "ì‹ì´ì„¬ìœ ": "ì•½ë¬¼ í¡ìˆ˜ ë°©í•´ ë° íš¨ê³¼ ì§€ì—°",
    "ì¹´í˜ì¸": "ìœ„ì‚° ì—­ë¥˜ ë° ìœ„ì¥ê´€ ì¶œí˜ˆ ìœ„í—˜",
    "ë§ˆëŠ˜": "í˜ˆì•¡ ì‘ê³  ë°©í•´ë¡œ ì¸í•œ ì¶œí˜ˆ ìœ„í—˜",
    "ìƒê°•": "ì§€í˜ˆ ë°©í•´ ë° ì¶œí˜ˆ ìœ„í—˜ ì¦ê°€",
    "ì¸ì‚¼": "ì¶œí˜ˆ ìœ„í—˜ì„±ì„ ì¦ê°€ì‹œí‚¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤.",
    "ì€í–‰ì—½": "ì§€í˜ˆ ë°©í•´ë¡œ ì¸í•œ ì¶œí˜ˆ ìœ„í—˜"
}

food_inference = {
    "ë¹„ë¹”ë°¥": ["ì‹ì´ì„¬ìœ ", "ë§ˆëŠ˜"], "ë‚˜ë¬¼": ["ì‹ì´ì„¬ìœ ", "ë§ˆëŠ˜"],
    "ìƒëŸ¬ë“œ": ["ì‹ì´ì„¬ìœ ", "ë¹„íƒ€ë¯¼C"], "ì¹˜í‚¨": ["ë§ˆëŠ˜"], "í†µë‹­": ["ë§ˆëŠ˜"],
    "ì•„ë©”ë¦¬ì¹´ë…¸": ["ì¹´í˜ì¸"], "ì»¤í”¼": ["ì¹´í˜ì¸"], "ì´ˆì½œë¦¿": ["ì¹´í˜ì¸"],
    "ì¼€ì´í¬": ["ì¹´í˜ì¸", "ì‹œë¦¬ì–¼"], "ì½œë¼": ["ì¹´í˜ì¸"], "ì—ë„ˆì§€ë“œë§í¬": ["ì¹´í˜ì¸"]
}

# 2. ë¶„ì„ ë¡œì§
def analyze_meal(meal_input):
    user_menus = [m.strip() for m in meal_input.split(',') if m.strip()]
    final_report = {}
    for menu in user_menus:
        detected = set()
        for key in reasons.keys():
            if key in menu.replace(" ", ""): detected.add(key)
        for fk, items in food_inference.items():
            if fk in menu:
                for itm in items: detected.add(itm)

        if detected:
            warnings, unsafe = [], set()
            for ing in detected:
                for drug, info in drug_db.items():
                    if ing in info["ì£¼ì˜ì‹ì¬ë£Œ"]:
                        warnings.append({"name": drug, "reason": reasons[ing], "brand": ", ".join(info["ìƒí’ˆëª…"])})
                        unsafe.add(drug)
            safe = [f"<b>{d}</b>({', '.join(drug_db[d]['ìƒí’ˆëª…'])})" for d in drug_db if d not in unsafe]
            final_report[menu] = {"detected": list(detected), "warnings": warnings, "safe": safe}
    return final_report

# 3. UI ë° ë””ìì¸ (í•˜ëŠ˜ìƒ‰ í…Œë§ˆ & ê·€ì—¬ìš´ í°íŠ¸ ìŠ¤íƒ€ì¼)
style_html = """
<style>
    @import url('https://fonts.googleapis.com/css2?family=Nanum+Gothic+Coding&family=Gaegu&display=swap');
    .main-container { font-family: 'Gaegu', cursive, sans-serif; background-color: #f0f9ff; padding: 20px; border-radius: 20px; }
    .header { background: #b3e5fc; padding: 25px; border-radius: 20px; text-align: center; color: #01579b; border: 3px dashed #81d4fa; }
    .card { background: white; border-radius: 15px; padding: 20px; margin-top: 20px; border: 2px solid #e1f5fe; box-shadow: 5px 5px 0px #b3e5fc; }
    .warn-item { background: #fff5f5; border-radius: 10px; padding: 10px; margin: 8px 0; border: 1px solid #ffcdd2; }
    .safe-box { background: #e8f5e9; padding: 12px; border-radius: 10px; color: #2e7d32; font-weight: bold; border: 1px solid #c8e6c9; text-align: center; }
    .danger-box { background: #ffebee; padding: 12px; border-radius: 10px; color: #c62828; font-weight: bold; border: 1px solid #ffcdd2; text-align: center; }
</style>
"""

header_ui = HTML(style_html + """
    <div class="header">
        <h1 style="margin: 0; font-size: 32px;">ğŸ’Š ê¸‰ì‹-ì§„í†µì œ ê¶í•© ê°€ì´ë“œ</h1>
        <p style="margin: 5px 0 0 0; font-size: 18px;">ì˜¤ëŠ˜ ë¨¹ì€ ê¸‰ì‹, ë‚´ ì•½ê³¼ ì˜ ë§ì„ê¹Œ? í™•ì¸í•´ë´!</p>
    </div>
""")

input_text = widgets.Text(placeholder='ë©”ë‰´ë¥¼ ì…ë ¥í•´ì¤˜! (ì˜ˆ: ë§ˆëŠ˜ í†µë‹­êµ¬ì´, ì´ˆì½œë¦¿ ì¼€ì´í¬)', layout={'width': '75%', 'height': '45px'})
btn = widgets.Button(description="ê¶í•© í™•ì¸í•˜ê¸° âœ¨", layout={'width': '23%', 'height': '45px'})
btn.style.button_color = '#4fc3f7'
btn.style.font_weight = 'bold'

output = widgets.Output()

def on_click(b):
    with output:
        clear_output()
        report = analyze_meal(input_text.value)
        if not report:
            display(HTML("<div style='text-align:center; padding:20px; font-size:18px; color:#555;'>ë©”ë‰´ë¥¼ ì‰¼í‘œ(,)ë¡œ êµ¬ë¶„í•´ì„œ ì…ë ¥í•´ì¤˜! ğŸ˜Š</div>"))
            return

        for menu, data in report.items():
            warn_html = ""
            seen = set()
            for w in data['warnings']:
                if w['name'] not in seen:
                    warn_html += f"""
                    <div class="warn-item">
                        <b style="color:#d32f2f;">ğŸ“ {w['name']} ì£¼ì˜!</b><br>
                        <small>ì•½ ì´ë¦„: {w['brand']}</small><br>
                        <div style="margin-top:5px;">ğŸ’¡ {w['reason']}</div>
                    </div>"""
                    seen.add(w['name'])

            status = f"""<div class="safe-box">âœ… ì¶”ì²œ ì„±ë¶„: {', '.join(data['safe'])}</div>""" if data['safe'] else """<div class="danger-box">ğŸš¨ ì£¼ì˜: ì§€ê¸ˆ ì‹ë‹¨ì—” ë§ëŠ” ì•½ì´ ì—†ì–´! <br> ê¼­ ì•½ì‚¬ë‹˜ê³¼ ìƒë‹´í•´ì•¼ í•´!</div>"""

            display(HTML(f"""
                <div class="card">
                    <h3 style="color:#0277bd; margin-top:0;">ğŸ½ï¸ ë©”ë‰´: {menu}</h3>
                    <p style="font-size:15px; color:#666;">ì°¾ì€ ì¬ë£Œ: <b>{', '.join(data['detected'])}</b></p>
                    {warn_html}
                    <div style="margin-top:15px;">{status}</div>
                </div>
            """))

btn.on_click(on_click)
display(header_ui, widgets.HBox([input_text, btn], layout={'margin': '20px 0'}), output)
```
